```python
np.random.seed(0)

temperature = 15 + 8 * np.random.randn(2, 3, 4)

precipitation = 10 * np.random.rand(2, 3, 4)

lon = [-99.83, -99.32]

lat = [42.25, 42.21]

instruments = ["manufac1", "manufac2", "manufac3"]

time = pd.date_range("2014-09-06", periods=4)

reference_time = pd.Timestamp("2014-09-05")

ds = xr.Dataset( {
        "temperature": (["loc", "instrument", "time"], temperature),
        "precipitation": (["loc", "instrument", "time"], precipitation),
    },
    coords={
        "lon": (["loc"], lon),
        "lat": (["loc"], lat),
        "instrument": instruments,
        "time": time,
        "reference_time": reference_time,
    },
)

ds
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Générer les données de température
temperature = 15 + 8 * np.random.randn(2, 3, 4)
# Créer la figure et les sous-graphiques
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
for i in range(2):
    ax = axes[i]
    cax = ax.matshow(temperature[i], cmap='coolwarm')
    fig.colorbar(cax, ax=ax)
    ax.set_title(f'Bloc {i+1}')

plt.show()
```
